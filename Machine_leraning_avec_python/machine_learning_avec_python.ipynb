 ### Qu'est ce que la machine learning 

le machine learning (Ml) est une  discipline de l'intellignece artificielle (IA) qui offre aux ordinateurs la possibilité d'apprendre à partir d'un ensemble d'obsmervations que l'on appel ensemble d'apprentissage

*classification des images, texts, vidéos, son
les publicités ciblées
diagnostic médical: quelles sont les maladies susceptibles de toucher une personnes

systèmes de recommendation 

les algorithme de machine learning peuvent être groupés en trois catégories :
    *Apprentissage supervisé: si les classes sont prédéterminées, ex: classer des bananes et les pommes
    *Apprentissage non supervisé: quand le système ou l'opérateur ne dispose pas d'étiquette, ex: regrouper les profils Youtube.
    *Apprentissage par renforcement: l'algorithme apprend un comportement étant donné une observation , ex: jeux vidéos 
        

### Principe genéral

* l'objectif des algorithmes du machine learning est de trouver la fonction cile f qui mappe le mieux les variables d'entrée x à une variable de sortie y= f(x)
* dans ce cas nous pourrons faire faire des prévisions dans l'avenir y à partir de nouveaux exemples de variabls d'entrée (x)
* un algorithme ne peut pas résoudre yous les problème de machine learning 

Pour définir les variables prédicives et les variables cibles,
commment par exemple on peut estimeer le prix d'une maison.
on a besoin de savoir le nombre de chiffre et le nombre d'écoles dans cette zone, l'emplacement, le taux de criminalité 

 * Les variables prédictives (ou attributs ou caractéristiques), dans notre exemple superficie, nombre de chambre, adresse, date de construction 
 * les variables cibles sont les variables à partir desquelles on espère pouvoir faire des prévisions. les n variables précédentesassociées à une observation seront notées comme un vecteur x=(x1, ...., xn) à n composantes

Régression et classification 
* de manièr egénérale, nous aimerons prédire une valeur y à partir dune observation x:
    y=f(x)
    * si y est continue : Regression 
    * si y est discret : classification 
    

### K-fold Cross Validation 

* Une autre technique pour protéger votre modèle contre l'overfitting est: k-fold cross validation 
* Principe: la séparation des données en K parties. La partie <test> change itérativement et on mesure les performances. En fin, on prend la moyenne des metrics sur toutes les itérations

## l'algorithme de K-means clustering

### principe de K-Means
* K-means clustering permet de diviser les données en k groupes 
#### principe:
* en réalité, comment cela fonctionne est vraiment simple:
    * Choisissez au hasard k centroides (k-moyennes)
    * Attribuez chaque point de données au centroide dont le plus proche 
    * Recalculez la distance moyenne entre centroides les points associés 
    * Itérez jusqu'à ce que vous aurez une moyenne optimale des distances 
    * Enfin, si vous voulez prédire le cluster pour de nouveaux points, trouvez simplement le centroide dont ils sont les plus proches.

On ne garde que l'information  volume du colis en fonction de poids de colis 

- la droite retenue (hyperplan) est celle dont la distance d (qui la sépare de tous les points au minimum) est maximale

- SVM permet donc de séparer les deux groupes en gardant le plus de marge possible.
- Lorsque l'on a plus de 2 classes, SVM(on parle de SVM multiclasse) fonctionne très bien

# l'écart type et la varience:
* le mode, la médiane et la moyenne nous donne une idée sur la tendance cetrale de la dataset o des variables
* Parcontre, la varience, nous donne une idée sur la disperssion de nos données, 
* La variance est simplement la moyenne des diférences au carré de la moyenne.
    - Exemple: Quelle est la varience de l'ensemble de données(1,4,5,4,8)
        * trouver la moyenne: (1+4+5+4+8)/5=4.4
        * trouver la différence par rapport à la moyenne: (-3.4, -0.4, 0.6, -0.4, 3.6)
        * trouver les différences au carrés: (11.56, 0.16, 0.36, 0.16, 12.96)
        * la varience = (11.56+0.16+0.36+0.16+12.96)/5=5.04
* l'écart type: c'est la racine carrée de la varience, il set à mesurer la dispersion d'une série statistique autour de sa moyenne
- plus l'écart-type est petit, plus les caractéristiques sont concentrés autour de la moyenne(on dit que la série est homogène).
- plus il est grand, plus les cractères sont dispersés autour de la moyenne(on dit que la série est hétérogène).
        * l'écart-type= 2.24

```python
class client(utilisateur):
    def __init__(self, nom, prenom, age):
        self.nom=nom
        self.prenom=prenom
        self.age=age
        self.solde=0
    def mon_solde(self,solde):
        self.solde=solde
    def Hello(self):
        return f"Hello, my name is {self.nom} {self.prenom} mon solde est de {self.solde}"
```

```python
client1=client('Mohamed', 'Said', 25)
```

```python
client1.mon_solde(400)
```

```python
client1.Hello()
```
